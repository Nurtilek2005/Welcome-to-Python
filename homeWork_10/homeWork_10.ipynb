```python
import pandas as pd
import random

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})

unique_vals = data['whoAmI'].unique()
one_hot = pd.DataFrame(0, index=data.index, columns=unique_vals)

for val in unique_vals:
    one_hot.loc[data['whoAmI'] == val, val] = 1

data = pd.concat([data, one_hot], axis=1)
data = data.drop('whoAmI', axis=1)
data
```
